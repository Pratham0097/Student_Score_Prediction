```python
import tkinter as tk
from tkinter import messagebox
import joblib
import numpy as np
```

```python
# Load the trained model
model = joblib.load("performance_predictor1211.joblib")
```

```python
# Create the main window
app = tk.Tk()
app.title("Performance Predictor")
app.geometry("350x200")
app.resizable(False, False)
```

```python
# Input label and entry
tk.Label(app, text="Hours Studied:").pack()
hours_entry = tk.Entry(app)
hours_entry.pack()

tk.Label(app, text="Attendance Rate (%):").pack()
attendance_entry = tk.Entry(app)
attendance_entry.pack()
```

```python
# function
def predict():
    try:
        hours = float(hours_entry.get())
        attendance = float(attendance_entry.get())
        result = model.predict([[hours, attendance]])
        messagebox.showinfo("Result", f"Estimated Final Score: {result[0]}%")
    except:
        messagebox.showerror("Error", "Please enter a valid number.")
```

```python
# Predict button
tk.Button(app, text="Predict", command=predict).pack()
```

```python
app.mainloop()
```
